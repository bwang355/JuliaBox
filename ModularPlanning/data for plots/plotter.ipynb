```julia
using LinearAlgebra, Statistics, CSV, DataFrames, DelimitedFiles
using PyPlot
```

```julia
a=readdlm("case_1_single.txt", ',', Float64)
b=readdlm("case_2_single.txt", ',', Float64)
c=readdlm("case_3_single.txt", ',', Float64)
d=readdlm("case_1_single_npv.txt", ',', Float64)
e=readdlm("case_2_single_npv.txt", ',', Float64)
f=readdlm("case_3_single_npv.txt", ',', Float64)
```

```julia
plot(a[:,1],a[:,2], label="case 1", color=:red)
plot(b[:,1],b[:,2], label="case 2", color=:black)
plot(c[:,1],c[:,2], label="case 3", color=:blue)
axis([0, 110000, 0, 100000])
grid(true)
ticklabel_format(axis="x", style="sci", scilimits=(0,0))
ticklabel_format(axis="y", style="sci", scilimits=(5,4))
legend(loc=4, fontsize=12)
xlabel("Risk(\$)", fontsize=14)
ylabel("Expected Value(\$)", fontsize=14)
xticks(0:22000:110000, fontsize=14)
yticks(0:10000:100000, fontsize=14)
savefig("single_result_1.eps")
```

```julia
plot(d[:,1],d[:,2], label="case 1", color=:red)
plot(e[:,1],e[:,2], label="case 2", color=:black)
plot(f[:,1],f[:,2], label="case 3", color=:blue)
axis([0, 110000, 0, 100000])
grid(true)
ticklabel_format(axis="x", style="sci", scilimits=(0,0))
ticklabel_format(axis="y", style="sci", scilimits=(4,5))
legend(loc=4, fontsize=12)
xlabel("Risk(\$)", fontsize=14)
ylabel("Expected Value(\$)", fontsize=14)
xticks(0:22000:110000, fontsize=14)
yticks(0:10000:100000, fontsize=14)
savefig("single_result_2.eps")
```

```julia
plot(a[:,1],a[:,2], color=:black)
plot(b[:,1],b[:,2], color=:black)
plot(c[:,1],c[:,2], label="cumulative problem", color=:black)
plot(d[:,1],d[:,2], color=:red)
plot(e[:,1],e[:,2],  color=:red)
plot(f[:,1],f[:,2], label="NPV problem", color=:red)
axis([0, 110000, 0, 100000])
grid(true)
ticklabel_format(axis="x", style="sci", scilimits=(0,0))
ticklabel_format(axis="y", style="sci", scilimits=(4,5))
legend(loc=4, fontsize=12)
xlabel("Risk(\$)", fontsize=14)
ylabel("Expected Value(\$)", fontsize=14)
xticks(0:22000:110000, fontsize=14)
yticks(0:10000:100000, fontsize=14)
savefig("case_single_stack.eps")
```
