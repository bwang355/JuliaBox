```julia
μ_c=10000;
m=Model(optimizer_with_attributes(Gurobi.Optimizer))

@variable(m, u[t in D, j in S[t], i in I, k in 1:N[i]] >= 0, Int)
#selection of plants with different capacity
@expression(m, x[t in D,j in S[t], i in I], sum(u[t,j,i,k]*B[i][k] for k in 1:N[i]))
@expression(m, y[t in D,j in S[t], i in I], sum(u[t,j,i,k]*C[i][k] for k in 1:N[i]))

# storage and waste
@variable(m, 0 <= s[t in T, j in S[t], i in I], Int)
@variable(m, 0 <= w[t in T, j in S[t], i in I], Int)
@constraint(m, [t in D, j in S[t], i in I], s[t,j,i] <= s_bar[i])

#cumulative variables
@variable(m, X[t in T, j in S[t], i in I] <= x_bar[i])
@variable(m, Y[t in D, j in S[t], i in I])

# cumulative constraints
@constraint(m, [i in I], X[1,S[1],i] == 0)
@constraint(m, [t in T[2:end], j in S[t], i in I], X[t,j,i] == X[a[t,j][1], a[t,j][2], i] + x[a[t,j][1], a[t,j][2], i])
@constraint(m, [i in I], Y[1,S[1],i] == y[1,S[1],i])
@constraint(m, [t in D[2:end], j in S[t], i in I], Y[t,j,i] == Y[a[t,j][1], a[t,j][2], i] + y[t,j,i])

#total fixed cost
@constraint(m, [j in S[T[end]], i in I], sum(Y[a[T[end],j][1], a[T[end],j][2],i] for i in I) <= y_bar)

# dependency of A and B
@constraint(m, [i in I, ii in I, t in D[2:end], j in S[t]], α[i][ii]*X[t,j,i] <= s[a[t,j][1], a[t,j][2], ii] + X[t,j,ii])

#storing
@constraint(m, [i in I], s[1,S[1],i] == 0)
@constraint(m, [i in I, j in S[T[end]]], s[T[end],j,i] == 0)
@constraint(m, [i in I], w[1,S[1],i] == 0)
@constraint(m, [t in T[2:end], j in S[t], i in I], 0<=X[t, j, i] + s[a[t,j][1], a[t,j][2], i] - s[t,j,i] - w[t,j,i] - sum(X[t, j, ii]*α[ii][i] for ii in I) <= d[i][t][j])

#cost
@variable(m, q[t in T, j in S[t]])
@constraint(m, [t in D, j in S[t]], q[t,j]==sum(y[t,j,i]+ρ_s[i]*s[t,j,i]+ρ_p[i] * X[t,j,i]+ρ_w[i]*w[t,j,i] for i in I))
@constraint(m, [j in S[T[end]]], q[T[end],j]==sum(ρ_w[i]*w[T[end],j,i]+ρ_p[i] * X[T[end],j,i] for i in I))

#revenue
@variable(m, r[t in T, j in S[t]])
@constraint(m, [j in S[1]], r[1,j]==0)
@constraint(m, [t in T[2:end], j in S[t]], r[t,j]==sum(π_p[i] * (s[a[t,j][1], a[t,j][2],i]+X[t,j,i] - sum(X[t,j,ii]*α[ii][i] for ii in I) - s[t,j,i] - w[t,j,i]) for i in I))

#discounting
@expression(m, β[t in T], 1 / (1 + γ) ^ (t - 1))

#mean absolute deviation variables
@expression(m, v[t in T, j in S[t]], β[t]*(r[t,j]-q[t,j]))
@variable(m, V[t in T, j in S[t]])
@constraint(m, V[1,S[1]]==v[1,S[1]])
@constraint(m, [t in T[2:end], j in S[t]], V[t,j]==V[a[t,j][1], a[t,j][2]]+v[t,j])
@variable(m, R[T[end], j in S[T[end]]] >= 0)
@expression(m, ϵ, sum(p[T[end]][j]*V[T[end],j] for j in S[T[end]]))

#mean_profit
@constraint(m, ϵ >= μ_c)

#Mean absolute deviation
@constraint(m, [j in S[T[end]]], R[T[end],j] >= p[T[end]][j]*(V[T[end],j] - ϵ))
@constraint(m, [j in S[T[end]]], R[T[end],j] >= p[T[end]][j]*(ϵ - V[T[end],j]))

@objective(m, Min, sum(R[T[end],j] for j in S[T[end]]))

status=optimize!(m)
```
